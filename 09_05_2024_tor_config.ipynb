```python
import scrapy
import scrapydo
from fake_useragent import UserAgent
from duckduckgo_search import DDGS

ua = UserAgent()
scrapydo.setup()

class ZoominfooSpider(scrapy.Spider):
    name = 'zoominfo'

    custom_settings = {
        'REQUEST_FINGERPRINTER_IMPLEMENTATION': '2.7',
        'USER_AGENT' : ua.random,
        'ROBOTSTXT_OBEY' : False,
        'CONCURRENT_REQUESTS_PER_DOMAIN': 1,  # Limit concurrent requests to avoid server overload
        'DOWNLOAD_DELAY': 5,
        'RETRY_TIMES': 3,
        'ROTATING_PROXY_LIST_PATH' : 'proxies.txt',
        'NUMBER_OF_PROXIES_TO_FETCH' : 10,
        'DOWNLOAD_TIMEOUT': 60,  # Increase timeout to 60 seconds
        'DOWNLOAD_FAIL_ON_DATALOSS': False
    }
    def start_requests(self):
        proxies = {
        'http': 'http://127.0.0.1:8118',
        'https': 'https://127.0.0.1:8118',
    }
        with open("input.csv", 'r') as input_file:
            for company_name in input_file:
                company = company_name.strip()
                url = f"https://www.google.com.ua/search?q={company + '+zoominfo+overview'}"
                # results = DDGS().text(f"{company}+zoominfooverview site:zoominfo.com inurl:{company}", max_results=1)
                # url = results[0]['href']
                req = scrapy.Request(url=url, callback=self.parse_google_results, cb_kwargs={"company": company},dont_filter=True)
                req.meta['proxy'] = proxies
                yield req

    def parse_google_results(self, response, **kwargs):
        all_links = response.css("a::attr(href)").getall()
        zoomlinks = [link for link in all_links if "www.zoominfo.com/c/" in link]
        print(zoomlinks[0])
        yield scrapy.Request(url=zoomlinks[0], callback=self.parse, cb_kwargs=kwargs)

    def parse(self, response, **kwargs):
        yield {
            'company': kwargs['company'],
            'headquarters': response.css('body > app-root > app-company > main > app-top-data > app-about > div:nth-child(2) > div > span::text').get(),
            'phone': response.css("body > app-root > app-company > main > app-top-data > app-about > div:nth-child(3) > div > span::text").get(),
            'revenue': response.css("body > app-root > app-company > main > app-top-data > app-about > div:nth-child(5) > div > span::text").get(),
            'employees_num': response.css("body > app-root > app-company > main > app-company-header > div.company-content > p.company-header-subtitle").get('href'),
            'website': response.css("body > app-root > app-company > main > app-top-data > app-about > div:nth-child(4) > div > a::text").get()
        }
```

```python
items = scrapydo.run_spider(ZoominfooSpider)
print(items)
```
